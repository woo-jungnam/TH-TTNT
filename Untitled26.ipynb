### Kiến trúc & Chiến lược:
1. **Backbone:** ResNet50 (Pre-trained on ImageNet). Lựa chọn ResNet50 vì cơ chế **Identity Mapping** giúp model học được các feature phức tạp mà không bị bão hòa độ chính xác.
2. **Regularization:** Sử dụng Dropout (0.5), BatchNormalization và Heavy Data Augmentation để chống Overfitting.
3. **Training Strategy:**
    - Phase 1: Freeze base model, train Head layers với LR cao.
    - Phase 2: Unfreeze top layers, fine-tuning với LR cực thấp để hội tụ tối ưu.
4. **Production Ready:** Tích hợp Callback (EarlyStopping, ReduceLROnPlateau) và Pipeline Inference riêng biệt.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
from tensorflow.keras.applications import ResNet50
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Cấu hình Global cho Production
CONFIG = {
    'IMG_SIZE': (224, 224),
    'BATCH_SIZE': 32,
    'EPOCHS_PHASE_1': 10,
    'EPOCHS_PHASE_2': 20,
    'LR_PHASE_1': 1e-3,
    'LR_PHASE_2': 1e-5,
    'VAL_SPLIT': 0.2,
    'CLASSES': 10  # Giả sử 10 lớp cho demo, Food-101 có 101 lớp
}

print(f"TensorFlow version: {tf.__version__}")
print("GPU is", "available" if tf.config.list_physical_devices('GPU') else "NOT AVAILABLE")
```

### 1. Data Pipeline & Augmentation
Chúng ta sử dụng `ImageDataGenerator` (hoặc Keras Preprocessing Layers) với các kỹ thuật: Random Rotation, Zoom, Brightness/Contrast adjustment để mô phỏng điều kiện thực tế (ảnh điện thoại, ánh sáng yếu).

```python
def get_data_pipelines():
    # Augmentation mạnh để chống overfitting và bias
    train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
        rescale=1./255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        brightness_range=[0.8, 1.2],
        fill_mode='nearest',
        validation_split=CONFIG['VAL_SPLIT']
    )

    test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

    return train_datagen, test_datagen

# Lưu ý: Ở bước này trong thực tế bạn sẽ trỏ path tới folder Food-101
print("Pipeline data đã sẵn sàng.")
```

### 1.1. Thực thi Loading Dataset (Food-101)
Sử dụng `tensorflow_datasets` để tải dữ liệu chuẩn, thực hiện split và chuẩn bị pipeline hiệu năng cao với `tf.data`.

```python
import tensorflow_datasets as tfds

def load_food101():
    print("Đang tải dataset Food-101... (Có thể mất vài phút)")
    (ds_train, ds_val, ds_test), ds_info = tfds.load(
        'food101',
        split=['train[:80%]', 'train[80%:]', 'validation'],
        shuffle_files=True,
        as_supervised=True,
        with_info=True
    )

    def preprocess_img(image, label):
        image = tf.image.resize(image, CONFIG['IMG_SIZE'])
        return tf.cast(image, tf.float32) / 255.0, label

    # Pipeline tối ưu cho GPU
    train_ds = ds_train.map(preprocess_img, num_parallel_calls=tf.data.AUTOTUNE)
    train_ds = train_ds.cache().shuffle(1000).batch(CONFIG['BATCH_SIZE']).prefetch(tf.data.AUTOTUNE)

    val_ds = ds_val.map(preprocess_img, num_parallel_calls=tf.data.AUTOTUNE)
    val_ds = val_ds.batch(CONFIG['BATCH_SIZE']).cache().prefetch(tf.data.AUTOTUNE)

    return train_ds, val_ds, ds_info

# train_ds, val_ds, info = load_food101()
# class_names = info.features['label'].names
print("Hàm load dữ liệu đã sẵn sàng.")
```

### 1.2. Chống Bias Dataset (Class Weights)
Nếu một số món ăn có ít dữ liệu hơn, model sẽ bị bias. Chúng ta sử dụng `compute_class_weight` để cân bằng lại tầm quan trọng của các lớp trong quá trình học.

```python
from sklearn.utils import class_weight

def get_class_weights(train_ds, info):
    labels = []
    for _, label in train_ds.unbatch():
        labels.append(label.numpy())

    weights = class_weight.compute_class_weight(
        class_weight='balanced',
        classes=np.unique(labels),
        y=labels
    )
    return dict(enumerate(weights))

print("Logic xử lý Class Imbalance đã sẵn sàng.")
```

### 2. Model Building (Transfer Learning)
Sử dụng ResNet50 với GlobalAveragePooling2D để giảm tham số, tránh Overfitting thay vì dùng các lớp Dense quá lớn.

```python
def build_model(num_classes):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(*CONFIG['IMG_SIZE'], 3))
    base_model.trainable = False  # Freeze lúc đầu

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(512, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax') # Softmax cho multi-class
    ])

    return model, base_model

model, base_model = build_model(CONFIG['CLASSES'])
model.compile(optimizer=optimizers.Adam(learning_rate=CONFIG['LR_PHASE_1']),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

### 3. Training Callbacks
Thiết lập các chốt chặn để bảo vệ model: `EarlyStopping` ngừng khi không còn tiến triển, `ModelCheckpoint` lưu model tốt nhất.

```python
checkpoint_path = "best_food_model.keras"

my_callbacks = [
    callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-7, verbose=1),
    callbacks.ModelCheckpoint(filepath=checkpoint_path, monitor='val_accuracy', save_best_only=True, verbose=1)
]

print("Callbacks đã được cấu hình.")
```

### 4. Training Phase 1: Training the Head
Trong giai đoạn này, chúng ta giữ nguyên các feature của ImageNet và chỉ tối ưu hóa các lớp phân loại phía trên.

### 4.1. Giải thích về Residual Connections
Lý do tôi chọn **ResNet50** là vì các **Skip Connections**. Thay vì ép các layer học toàn bộ ánh xạ $H(x)$, chúng ta ép chúng học phần dư (residual) $F(x) = H(x) - x$. Điều này giúp đạo hàm có thể 'chảy' ngược về các layer đầu mà không bị triệt tiêu, giúp model hội tụ cực nhanh và sâu.

### 5. Training Phase 2: Fine-tuning Base Model
Sau khi Head đã ổn định, chúng ta unfreeze khoảng 20-30 layers cuối của ResNet50 để tinh chỉnh các feature chuyên biệt cho món ăn.

```python
base_model.trainable = True

# Unfreeze các layer cuối (thường là từ layer 143 trở đi của ResNet50)
# Chúng ta giữ lại các layer đầu (low-level features như cạnh, màu sắc) và chỉ train layer sâu hơn
for layer in base_model.layers[:-20]:
    layer.trainable = False

model.compile(optimizer=optimizers.Adam(learning_rate=CONFIG['LR_PHASE_2']),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

print("--- Starting Phase 2: Fine-tuning Base Layers ---")
# history_phase2 = model.fit(
#     train_generator,
#     epochs=CONFIG['EPOCHS_PHASE_2'],
#     validation_data=val_generator,
#     callbacks=my_callbacks
# )
```

### 5.1. Automated Training Orchestrator
Hàm này giúp chạy toàn bộ quy trình training một cách tự động và an toàn.

```python
def run_full_training(model, train_ds, val_ds, class_weights=None):
    print("--- GIAI ĐOẠN 1: TRAINING HEAD ---")
    history1 = model.fit(
        train_ds,
        epochs=CONFIG['EPOCHS_PHASE_1'],
        validation_data=val_ds,
        callbacks=my_callbacks,
        class_weight=class_weights
    )

    print("--- GIAI ĐOẠN 2: FINE-TUNING ---")
    # Unfreeze logic
    base_model.trainable = True
    for layer in base_model.layers[:-20]:
        layer.trainable = False

    model.compile(optimizer=optimizers.Adam(learning_rate=CONFIG['LR_PHASE_2']),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    history2 = model.fit(
        train_ds,
        epochs=CONFIG['EPOCHS_PHASE_2'],
        validation_data=val_ds,
        callbacks=my_callbacks,
        class_weight=class_weights
    )

    return history1, history2

print("Orchestrator đã sẵn sàng.")
```

### 10. Main Execution - Bắt đầu huấn luyện
Cell này kết hợp tất cả các module trên để thực hiện quy trình End-to-End: Load Data -> Tính Class Weights -> Chạy Orchestrator.

```python
# 1. Tải dữ liệu thực tế
train_ds, val_ds, info = load_food101()
class_names = info.features['label'].names
num_classes = info.features['label'].num_classes

# 2. Khởi tạo lại model với đúng số lượng lớp của Food-101 (101 lớp)
model, base_model = build_model(num_classes)
model.compile(optimizer=optimizers.Adam(learning_rate=CONFIG['LR_PHASE_1']),
              loss='sparse_categorical_crossentropy', # Dùng sparse vì nhãn từ tfds là integer
              metrics=['accuracy'])

# 3. Tính toán trọng số lớp
weights = get_class_weights(train_ds, info)

# 4. Thực thi huấn luyện
history_1, history_2 = run_full_training(model, train_ds, val_ds, class_weights=weights)

print("Huấn luyện hoàn tất!")
```

### 6. Professional Inference Pipeline
Sau khi train, model cần một pipeline dự đoán có độ tin cậy cao, bao gồm xử lý ảnh thô và hiển thị Confidence Score.

```python
def predict_and_visualize(img_path, model, class_names, threshold=0.5):
    # Load và preprocess ảnh thực tế
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=CONFIG['IMG_SIZE'])
    img_array = tf.keras.preprocessing.image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Dự đoán
    preds = model.predict(img_array)[0]
    top_idx = np.argsort(preds)[-3:][::-1] # Lấy Top 3

    # Visualization logic chuyên nghiệp
    plt.imshow(img)
    plt.axis('off')

    result_text = ""
    for i in top_idx:
        confidence = preds[i]
        label = class_names[i]
        result_text += f"{label}: {confidence*100:.2f}%\n"

    plt.title(result_text)
    plt.show()

    if preds[top_idx[0]] < threshold:
        print("Cảnh báo: Độ tự tin thấp, có thể là món ăn chưa có trong database.")

print("Inference Pipeline đã sẵn sàng.")
```

### 6.1. Giảm False Confidence (Calibration)
Trong Production 2025, chúng ta không tin mù quáng vào Softmax. Nếu `max(preds)` thấp hơn một ngưỡng (ví dụ 0.6), hệ thống nên báo cáo là 'Unknown' thay vì cố gắng phân loại bừa. Điều này giúp hệ thống hoạt động cực kỳ ổn định ngoài thực tế.

```python
def reliable_predict(img_path, model, class_names, threshold=0.65):
    # Tiền xử lý tương tự predict_and_visualize
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=CONFIG['IMG_SIZE'])
    img_array = tf.keras.preprocessing.image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    preds = model.predict(img_array)[0]
    max_conf = np.max(preds)
    predicted_class = np.argmax(preds)

    if max_conf < threshold:
        return "Unknown / Low Confidence", max_conf
    else:
        return class_names[predicted_class], max_conf

print("Pipeline Predict đáng tin cậy đã sẵn sàng.")
```

```python
def plot_performance(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs_range = range(len(acc))

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.title('Accuracy Performance')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.title('Loss Performance')
    plt.legend()
    plt.show()

# def detailed_report(model, val_gen):
#     Y_pred = model.predict(val_gen)
#     y_pred = np.argmax(Y_pred, axis=1)
#     print('Confusion Matrix')
#     print(confusion_matrix(val_gen.classes, y_pred))
#     print('Classification Report')
#     print(classification_report(val_gen.classes, y_pred, target_names=val_gen.class_indices.keys()))
```
