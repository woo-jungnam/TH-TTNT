```python
# ==============================================================================
# BƯỚC 1: CÀI ĐẶT CÁC THƯ VIỆN CẦN THIẾT
# ==============================================================================
!pip install -q datasets evaluation seaborn matplotlib

import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from datasets import load_dataset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import re
from tqdm import tqdm
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Đang chạy trên thiết bị: {device}")

# ==============================================================================
# BƯỚC 2: TẢI DATASET VÀ XỬ LÝ NHÃN (SỬA LỖI VALUEERROR: '53_')
# ==============================================================================
print("\n1. Đang tải dataset từ Hugging Face...")
raw_dataset = load_dataset("tiennv/utkface")

# Hàm bóc tách số tuổi an toàn từ chuỗi (Xử lý các case như '53_', '25_0_1')
def convert_age_to_group(example):
    raw_label = str(example['label']).strip()

    # Sử dụng Regex để trích xuất chuỗi số đầu tiên xuất hiện
    age_match = re.match(r'^(\d+)', raw_label)

    if age_match:
        age = int(age_match.group(1))
    else:
        age = 25 # Giá trị mặc định nếu chuỗi hoàn toàn không có số

    # Phân nhóm tuổi dựa trên số tuổi đã làm sạch
    if age <= 19:    group = 0 # Trẻ em / Teen
    elif age <= 35:  group = 1 # Người trẻ / Trưởng thành
    elif age <= 55:  group = 2 # Trung niên
    else:            group = 3 # Người già
    return {'age_group': group}

print("2. Đang phân loại dữ liệu thành các nhóm tuổi cho từng tập con...")
mapped_train_base = raw_dataset['train'].map(convert_age_to_group)
mapped_test_base = raw_dataset['test'].map(convert_age_to_group)

# Chia tập dữ liệu: Train 90% | Val 10%
split_train_val = mapped_train_base.train_test_split(test_size=0.1, seed=42)

train_data = split_train_val['train']
val_data = split_train_val['test']
test_data = mapped_test_base

# Lấy số chiều của Vector Embedding đầu vào
embedding_dim = len(train_data[0]['embd'])
train_labels = [example['age_group'] for example in train_data]
num_classes = len(np.unique(train_labels))

print(f" -> Cấu trúc dữ liệu: Kích thước Vector Đầu vào = {embedding_dim} | Số nhóm tuổi thực tế = {num_classes}")
print(f" -> Quy mô dữ liệu: Train ({len(train_data)}) | Val ({len(val_data)}) | Test ({len(test_data)})")

# ==============================================================================
# BƯỚC 3: TỐI ƯU CÂN BẰNG DỮ LIỆU (CLASS WEIGHTS)
# ==============================================================================
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_labels),
    y=np.array(train_labels)
)
class_weights = torch.FloatTensor(class_weights).to(device)
print(f" -> Trọng số tối ưu cho các class: {class_weights.cpu().numpy()}")

# ==============================================================================
# BƯỚC 4: ĐỊNH NGHĨA PYTORCH DATASET CHO VECTOR EMBEDDING
# ==============================================================================
class UTKFaceEmbeddingDataset(Dataset):
    def __init__(self, hf_dataset):
        self.dataset = hf_dataset

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        embedding = torch.tensor(item['embd'], dtype=torch.float32)
        label = item['age_group']
        return embedding, torch.tensor(label, dtype=torch.long)

# Khởi tạo DataLoaders
train_loader = DataLoader(UTKFaceEmbeddingDataset(train_data), batch_size=128, shuffle=True)
val_loader = DataLoader(UTKFaceEmbeddingDataset(val_data), batch_size=128, shuffle=False)
test_loader = DataLoader(UTKFaceEmbeddingDataset(test_data), batch_size=128, shuffle=False)

# ==============================================================================
# BƯỚC 5: KHỞI TẠO KIẾN TRÚC MẠNG MLP (MULTI-LAYER PERCEPTRON)
# ==============================================================================
class AgeGroupMLP(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(AgeGroupMLP, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(),
            nn.Dropout(0.4),

            nn.Linear(256, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.network(x)

model = AgeGroupMLP(input_dim=embedding_dim, num_classes=num_classes).to(device)

criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# ==============================================================================
# BƯỚC 6: VÒNG LẶP HUẤN LUYỆN (TRAINING LOOP)
# ==============================================================================
num_epochs = 15
best_val_loss = float('inf')

print("\n--- Bắt đầu quá trình huấn luyện mạng ---")
for epoch in range(num_epochs):
    # --- PHASE TRAIN ---
    model.train()
    running_loss, correct, total = 0.0, 0, 0

    train_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]")
    for embeddings, labels in train_bar:
        embeddings, labels = embeddings.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(embeddings)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * embeddings.size(0)
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

        train_bar.set_postfix(loss=loss.item(), acc=100.0 * correct / total)

    epoch_train_loss = running_loss / len(train_loader.dataset)
    epoch_train_acc = 100.0 * correct / total

    # --- PHASE VALIDATION ---
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0

    with torch.no_grad():
        for embeddings, labels in val_loader:
            embeddings, labels = embeddings.to(device), labels.to(device)
            outputs = model(embeddings)
            loss = criterion(outputs, labels)

            val_loss += loss.item() * embeddings.size(0)
            _, predicted = outputs.max(1)
            val_total += labels.size(0)
            val_correct += predicted.eq(labels).sum().item()

    epoch_val_loss = val_loss / len(val_loader.dataset)
    epoch_val_acc = 100.0 * val_correct / val_total

    print(f" ✨ Kết quả Epoch {epoch+1}: Train Loss: {epoch_train_loss:.4f} | Train Acc: {epoch_train_acc:.2f}% | Val Loss: {epoch_val_loss:.4f} | Val Acc: {epoch_val_acc:.2f}%")

    scheduler.step(epoch_val_loss)

    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        torch.save(model.state_dict(), 'best_embedding_model.pth')
        print(" -> Đã lưu cấu hình mạng tốt nhất!")

# ==============================================================================
# BƯỚC 7: ĐÁNH GIÁ TRÊN TẬP TEST ĐỘC LẬP
# ==============================================================================
print("\n--- Đang đánh giá tổng thể trên tập Test độc lập ---")
if os.path.exists('best_embedding_model.pth'):
    model.load_state_dict(torch.load('best_embedding_model.pth', map_location=device))
model.eval()

test_correct, test_total = 0, 0
with torch.no_grad():
    for embeddings, labels in tqdm(test_loader, desc="Testing"):
        embeddings, labels = embeddings.to(device), labels.to(device)
        outputs = model(embeddings)
        _, predicted = outputs.max(1)
        test_total += labels.size(0)
        test_correct += predicted.eq(labels).sum().item()

final_test_acc = 100.0 * test_correct / test_total
print(f"\n👉 [KẾT QUẢ] Độ chính xác (Accuracy) tổng thể trên tập Test: {final_test_acc:.2f}%")

# ==============================================================================
# BƯỚC 8: ĐÁNH GIÁ CHUYÊN SÂU (CONFUSION MATRIX & REPORT)
# ==============================================================================
print("\n--- Đang thu thập dự đoán chi tiết để lập báo cáo ---")
all_preds = []
all_labels = []

with torch.no_grad():
    for embeddings, labels in test_loader:
        embeddings = embeddings.to(device)
        outputs = model(embeddings)
        _, predicted = outputs.max(1)

        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.numpy())

# Khởi tạo nhãn hiển thị khớp động với số lớp thực tế có trong dữ liệu nhãn đã sạch
base_names = ['Nhom 0 (<=19)', 'Nhom 1 (20-35)', 'Nhom 2 (36-55)', 'Nhom 3 (>55)']
target_names = base_names[:num_classes]

print("\n[BÁO CÁO PHÂN LOẠI CHI TIẾT (PRECISION - RECALL - F1)]")
print(classification_report(all_labels, all_preds, target_names=target_names))

# Vẽ biểu đồ Ma trận nhầm lẫn
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=target_names, yticklabels=target_names)
plt.xlabel('Dự đoán từ Mô hình')
plt.ylabel('Nhãn Thực tế')
plt.title('Ma trận nhầm lẫn (Confusion Matrix)')
plt.show()
```

```python
# ==============================================================================
# BƯỚC 9: DỰ ĐOÁN THỬ NGHIỆM TRÊN MỘT MẪU NGẪU NHIÊN (INFERENCE TEST)
# ==============================================================================
print("\n--- Chạy thử nghiệm dự đoán trên một mẫu ngẫu nhiên từ tập Test ---")

# 1. Tải lại trọng số tốt nhất đã lưu
if os.path.exists('best_embedding_model.pth'):
    model.load_state_dict(torch.load('best_embedding_model.pth', map_location=device))
model.eval()

# 2. Lấy một mẫu ngẫu nhiên từ tập test gốc để kiểm tra
random_idx = random.randint(0, len(test_data) - 1)
test_sample = test_data[random_idx]

# Chuẩn bị dữ liệu đầu vào (Thêm batch dimension bằng unsqueeze(0))
test_embedding = torch.tensor(test_sample['embd'], dtype=torch.float32).unsqueeze(0).to(device)
true_age_raw = test_sample['label']
true_group_idx = test_sample['age_group']

# 3. Đưa qua mô hình dự đoán
with torch.no_grad():
    output = model(test_embedding)
    # Tính xác suất phần trăm cho từng nhóm bằng hàm Softmax
    probabilities = torch.nn.functional.softmax(output[0], dim=0).cpu().numpy()
    pred_group_idx = np.argmax(probabilities)

# 4. Hiển thị kết quả ra màn hình dạng text trực quan
print("=" * 60)
print(f" Mẫu test số: {random_idx}")
print(f" -> Tuổi gốc ghi trên nhãn: {true_age_raw}")
print(f" -> Nhóm tuổi thực tế: {target_names[true_group_idx]}")
print("-" * 60)
print(f" 🤖 MÔ HÌNH DỰ ĐOÁN: {target_names[pred_group_idx]} (Độ tin cậy: {probabilities[pred_group_idx]*100:.2f}%)")
print("=" * 60)

# 5. Vẽ biểu đồ cột hiển thị độ tin cậy của toàn bộ các nhóm tuổi
plt.figure(figsize=(7, 4))
colors = ['gray'] * num_classes
colors[pred_group_idx] = 'green' if pred_group_idx == true_group_idx else 'red' # Xanh nếu đúng, Đỏ nếu đoán sai

bars = plt.bar(target_names, probabilities * 100, color=colors, edgecolor='black', alpha=0.7)
plt.ylabel('Độ tin cậy (%)')
plt.title(f'Phân phối xác suất dự đoán của mô hình (Thực tế: {target_names[true_group_idx]})')
plt.ylim(0, 100)

# Ghi số phần trăm cụ thể lên đầu mỗi cột
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 2, f'{yval:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
